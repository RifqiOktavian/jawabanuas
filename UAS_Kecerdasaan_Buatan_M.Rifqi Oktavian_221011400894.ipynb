```python
import numpy as np
import matplotlib.pyplot as plt

class FuzzyHappinessPredictor:
    def __init__(self):
        pass

    def membership_pelayanan(self, x):
        return {
            'LAMBAT': max(0, min((5-x)/(5-0), 1)) if x <= 5 else 0,
            'SEDANG': max(0, min((x-3)/(7-3), (10-x)/(10-7))) if 3 <= x <= 10 else 0,
            'CEPAT': max(0, min((x-8)/(10-8), 1)) if x >= 8 else 0
        }

    def membership_makanan(self, x):
        return {
            'BURUK': max(0, min((5-x)/(5-0), 1)) if x <= 5 else 0,
            'SEDANG': max(0, min((x-3)/(7-3), (10-x)/(10-7))) if 3 <= x <= 10 else 0,
            'BAIK': max(0, min((x-8)/(10-8), 1)) if x >= 8 else 0
        }

    def membership_suasana(self, x):
        return {
            'TIDAK_NYAMAN': max(0, min((5-x)/(5-0), 1)) if x <= 5 else 0,
            'SEDANG': max(0, min((x-3)/(7-3), (10-x)/(10-7))) if 3 <= x <= 10 else 0,
            'NYAMAN': max(0, min((x-8)/(10-8), 1)) if x >= 8 else 0
        }

def plot_membership():
    x = np.linspace(0, 10, 100)

    predictor = FuzzyHappinessPredictor()

    pelayanan = {
        'LAMBAT': [predictor.membership_pelayanan(val)['LAMBAT'] for val in x],
        'SEDANG': [predictor.membership_pelayanan(val)['SEDANG'] for val in x],
        'CEPAT': [predictor.membership_pelayanan(val)['CEPAT'] for val in x],
    }

    makanan = {
        'BURUK': [predictor.membership_makanan(val)['BURUK'] for val in x],
        'SEDANG': [predictor.membership_makanan(val)['SEDANG'] for val in x],
        'BAIK': [predictor.membership_makanan(val)['BAIK'] for val in x],
    }

    suasana = {
        'TIDAK_NYAMAN': [predictor.membership_suasana(val)['TIDAK_NYAMAN'] for val in x],
        'SEDANG': [predictor.membership_suasana(val)['SEDANG'] for val in x],
        'NYAMAN': [predictor.membership_suasana(val)['NYAMAN'] for val in x],
    }

    plt.figure(figsize=(15, 10))

    plt.subplot(3, 1, 1)
    plt.plot(x, pelayanan['LAMBAT'], label='LAMBAT')
    plt.plot(x, pelayanan['SEDANG'], label='SEDANG')
    plt.plot(x, pelayanan['CEPAT'], label='CEPAT')
    plt.title('Fungsi Keanggotaan Pelayanan')
    plt.xlabel('Nilai')
    plt.ylabel('Derajat Keanggotaan')
    plt.legend()

    plt.subplot(3, 1, 2)
    plt.plot(x, makanan['BURUK'], label='BURUK')
    plt.plot(x, makanan['SEDANG'], label='SEDANG')
    plt.plot(x, makanan['BAIK'], label='BAIK')
    plt.title('Fungsi Keanggotaan Makanan')
    plt.xlabel('Nilai')
    plt.ylabel('Derajat Keanggotaan')
    plt.legend()

    plt.subplot(3, 1, 3)
    plt.plot(x, suasana['TIDAK_NYAMAN'], label='TIDAK_NYAMAN')
    plt.plot(x, suasana['SEDANG'], label='SEDANG')
    plt.plot(x, suasana['NYAMAN'], label='NYAMAN')
    plt.title('Fungsi Keanggotaan Suasana')
    plt.xlabel('Nilai')
    plt.ylabel('Derajat Keanggotaan')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_membership()
```
